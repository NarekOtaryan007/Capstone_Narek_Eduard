```python
from pathlib import Path
import os, shutil

fname = "free-exercise-db-main.zip"
p = Path(fname)

print("File:", p, "| size:", p.stat().st_size, "bytes")

# Read magic bytes (file signature)
with open(p, "rb") as f:
    head = f.read(16)
print("Magic bytes:", head)

out_dir = Path("free_exercise_extracted")
if out_dir.exists():
    shutil.rmtree(out_dir)
out_dir.mkdir(parents=True, exist_ok=True)

extracted_ok = False

# 1) Try ZIP
try:
    import zipfile
    with zipfile.ZipFile(p, "r") as z:
        z.extractall(out_dir)
    print("✅ Extracted as ZIP into:", out_dir)
    extracted_ok = True
except Exception as e:
    print("ZIP extract failed:", repr(e))

# 2) Try TAR / TAR.GZ / TGZ (if ZIP failed)
if not extracted_ok:
    try:
        import tarfile
        with tarfile.open(p, "r:*") as t:
            t.extractall(out_dir)
        print("✅ Extracted as TAR (auto-detected) into:", out_dir)
        extracted_ok = True
    except Exception as e:
        print("TAR extract failed:", repr(e))

# 3) Show what we extracted
print("\nExtracted files (first 200):")
all_files = [str(x) for x in out_dir.rglob("*") if x.is_file()]
for x in all_files[:200]:
    print(x)
print("\nTotal files:", len(all_files))

# 4) Find candidate data files
print("\nCandidate data files:")
candidates = []
for ext in (".json", ".csv", ".parquet", ".xlsx"):
    candidates += [str(x) for x in out_dir.rglob(f"*{ext}")]
for x in candidates[:200]:
    print(x)
print("\nTotal candidates:", len(candidates))
```

```python
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

# Folder with the per-exercise JSON files
ex_dir = Path("free_exercise_extracted/free-exercise-db-main/exercises")

# Load all exercises into a list of dicts
records = []
bad_files = []

for fp in ex_dir.glob("*.json"):
    try:
        with open(fp, "r", encoding="utf-8") as f:
            d = json.load(f)
        d["_file"] = fp.name
        records.append(d)
    except Exception as e:
        bad_files.append((fp.name, str(e)))

df = pd.DataFrame(records)

print("Loaded exercises:", len(df))
print("Bad files:", len(bad_files))
print("Columns:", df.columns.tolist())
display(df.head())

# ---- Helper for top counts (handles list columns too) ----
def top_counts(col, top_n=20):
    s = df[col]
    if s.apply(lambda x: isinstance(x, list)).any():
        s = s.explode()
    s = s.dropna().astype(str)
    return s.value_counts().head(top_n)

# ---- Visualizations (only if columns exist) ----
plot_cols = ["level", "category", "equipment", "force", "mechanic"]
for c in plot_cols:
    if c in df.columns:
        vc = top_counts(c, 20)
        plt.figure(figsize=(9,4))
        plt.title(f"Top values: {c}")
        plt.bar(vc.index, vc.values)
        plt.xticks(rotation=75, ha="right")
        plt.ylabel("Count")
        plt.tight_layout()
        plt.show()

# Muscles (list fields)
for c in ["primaryMuscles", "secondaryMuscles"]:
    if c in df.columns:
        vc = top_counts(c, 25)
        plt.figure(figsize=(10,4))
        plt.title(f"Top values: {c}")
        plt.bar(vc.index, vc.values)
        plt.xticks(rotation=75, ha="right")
        plt.ylabel("Count")
        plt.tight_layout()
        plt.show()

# Instruction steps distribution (if instructions is list)
if "instructions" in df.columns:
    steps = df["instructions"].apply(lambda x: len(x) if isinstance(x, list) else 0)
    plt.figure(figsize=(7,3.5))
    plt.title("Distribution: Instruction step count")
    plt.hist(steps, bins=30)
    plt.xlabel("Steps")
    plt.ylabel("Exercises")
    plt.tight_layout()
    plt.show()
    print("Instruction steps stats:")
    display(steps.describe())

# Show bad files (if any)
if bad_files:
    print("\nExamples of bad files:")
    for x in bad_files[:10]:
        print(x)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

display(df.head())
print("Rows:", len(df))
print("Columns:", df.columns.tolist())

# Missing values %
missing = df.isna().mean().sort_values(ascending=False)
missing = missing[missing > 0]
if len(missing) > 0:
    plt.figure(figsize=(10,4))
    plt.title("Missing rate per column")
    plt.bar(missing.index.astype(str), missing.values)
    plt.xticks(rotation=75, ha="right")
    plt.ylabel("Missing rate")
    plt.tight_layout()
    plt.show()
else:
    print("No missing values detected (NaN).")

def plot_top(col, top_n=20):
    s = df[col]
    if s.apply(lambda x: isinstance(x, list)).any():
        s = s.explode()
    s = s.fillna("Unknown").astype(str)
    vc = s.value_counts().head(top_n)
    plt.figure(figsize=(9,4))
    plt.title(f"Top values: {col}")
    plt.bar(vc.index, vc.values)
    plt.xticks(rotation=75, ha="right")
    plt.ylabel("Count")
    plt.tight_layout()
    plt.show()

# Categorical distributions
for c in ["level", "category", "equipment", "force", "mechanic"]:
    if c in df.columns:
        plot_top(c, 20)

# Muscle distributions
for c in ["primaryMuscles", "secondaryMuscles"]:
    if c in df.columns:
        plot_top(c, 25)

# Instruction steps distribution
steps = df["instructions"].apply(lambda x: len(x) if isinstance(x, list) else 0)
plt.figure(figsize=(7,3.5))
plt.title("Instruction step count distribution")
plt.hist(steps, bins=30)
plt.xlabel("Steps")
plt.ylabel("Exercises")
plt.tight_layout()
plt.show()

print("Instruction steps stats:")
display(steps.describe())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Build a flat table: one row = one primary muscle for an exercise
muscle_df = df[["id", "name", "category", "level", "equipment", "primaryMuscles"]].copy()
muscle_df = muscle_df.explode("primaryMuscles")
muscle_df["primaryMuscles"] = muscle_df["primaryMuscles"].fillna("Unknown").astype(str)
muscle_df["category"] = muscle_df["category"].fillna("Unknown").astype(str)

# Crosstab
ct = pd.crosstab(muscle_df["primaryMuscles"], muscle_df["category"])

plt.figure(figsize=(10,7))
plt.title("Exercises count: Primary muscle × Category")
plt.imshow(ct.values, aspect="auto")
plt.xticks(range(len(ct.columns)), ct.columns, rotation=75, ha="right")
plt.yticks(range(len(ct.index)), ct.index)
plt.colorbar(label="Count")
plt.tight_layout()
plt.show()

display(ct.sort_index())
```
